# Matrix multiplication

```python
mat_1=np.arange(1,7).reshape(3,2)
mat_1
```

```python
mat_2=np.arange(2,6).reshape(2,2)
mat_2
```

```python
mat_1@mat_2    # matrix multiplication = @
```

```python
np.dot(mat_1,mat_2) # dot matrix multiplication
```

```python
mat_1
```

```python
mat_1.T   # transpose
```

# stacking 

```python
# horizental
a=np.array([1,2,3])
b=np.array([4,5,6])
np.hstack((a,b))
```

```python
#vertical
np.vstack((a,b))
```

```python
# cloumn stack
np.column_stack((a,b))
```
